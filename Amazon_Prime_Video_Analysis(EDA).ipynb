## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# handling missing value
titles_df['description']=titles_df['description'].fillna('no description')
titles_df['age_certification']=titles_df['age_certification'].fillna('no certification')
titles_df['seasons']=titles_df['seasons'].fillna(0)
titles_df['imdb_score']=titles_df['imdb_score'].fillna(titles_df['imdb_score'].median())
titles_df['imdb_votes']=titles_df['imdb_votes'].fillna(titles_df['imdb_votes'].median())
titles_df['tmdb_popularity']=titles_df['tmdb_popularity'].fillna(titles_df['tmdb_popularity'].mean())
titles_df['tmdb_score']=titles_df['tmdb_score'].fillna(titles_df['tmdb_score'].mean())
credits_df['character']=credits_df['character'].fillna('no character')
# drop the duplicates
titles_df=titles_df.drop_duplicates()
# dopna the colums
titles_df=titles_df.dropna(subset=['age_certification'])
titles_df=titles_df.dropna(subset=['seasons'])
credits_df=credits_df.dropna(subset=['character'])
# Rename the colums
titles_df.rename(columns={'release_year':'year'},inplace=True)
credits_df.rename(columns={'role':'Role'},inplace=True)
# merging the datsets
merged_df=pd.merge(titles_df,credits_df,on='id',how='left')
merged_df=merged_df.drop_duplicates(subset=['title'])
top_popular=merged_df.sort_values(by='tmdb_popularity',ascending=False).head(10)['title']
high_rated_movies=merged_df[merged_df['imdb_score']>7].head(10)['title']
year_with_most_content=merged_df['year'].value_counts().idxmax()
yearly_distribution=merged_df['year'].value_counts().sort_index().tail()
content_count=merged_df.groupby('type')['id'].count()
avg_rating=merged_df.groupby('type')['imdb_score'].mean()
avg_score=merged_df.groupby('type')['tmdb_score'].mean()
top10_genres=merged_df['genres'].value_counts().head(10)
most_countires_content=merged_df['production_countries'].value_counts().head(10)
top_actors=merged_df.groupby('name')['id'].count().sort_values(ascending=False).head(10)
yearly_type_distribution=merged_df.groupby('year')['type'].value_counts().tail(10)
rating_year=merged_df.groupby('year')['imdb_score'].mean()
print(f'top popular movies:',top_popular)
print(f'high rated movies:',high_rated_movies)
print(f'year with most content:',year_with_most_content)
print(f'yearly distribution:',yearly_distribution)
print(f'content count:',content_count)
print(f'avg rating:',avg_rating)
print(f'avg score:',avg_score)
print(f'top10 genres:',top10_genres)
print(f'most countires content:',most_countires_content)
print(f'top actors:',top_actors)
print(f'yearly type distribution:',yearly_type_distribution)
print(f'rating year:',rating_year)
````
